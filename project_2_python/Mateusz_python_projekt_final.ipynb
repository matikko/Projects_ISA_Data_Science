```python
#import_bibliotek

import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Environment_Temperature_change_E_All_Data_NOFLAG.csv", encoding='cp1252') #wczoytanie ramki danych
```

```python
data #ramka danych
```

```python
# usuwanie non values, usuwanie Y z roku
clean_data=data.dropna()
clean_data.columns=clean_data.columns.str.replace('Y','')
```

```python
clean_data
```

```python
# usuwanie z danych standard deviation
data=clean_data[clean_data['Element']=='Temperature change']
# usuwanie niepotrzebnych kolumn
data=data.drop(columns=['Area Code','Months Code','Element Code'])
# bierzemy pod uwagę jednynie cały rok meteorologiczny
Temp_month=data.loc[data.Months.isin(['Meteorological year'])]
```

```python
# sprawdzenie
Temp_month
```

```python
# wypisanie wszystkich wartości z kolumny area w celu wybrania potrzebnych do mojej części projektowej
Temp_month.Area.unique()
```

```python
# wybranie obszarów do mojej części projektu
my_areas = [ 'Annex I countries', 'Land Locked Developing Countries', 'Least Developed Countries', 'Low Income Food Deficit Countries', 'Net Food Importing Developing Countries', 'Non-Annex I countries', 'OECD', 'Small Island Developing States', 'World']
```

```python
# zafiltrowanie danych do wybranych obszarów
my_data = Temp_month[Temp_month['Area'].isin(my_areas)]
```

```python
# sprawdzenie
my_data
```

```python
# opisanie średnie zmiany temp w całym roku w danej grupie area
my_data_avg = my_data.describe()
my_data_avg
```

```python
# średnia wartość zmiany temp w danym roku
avg_temp = my_data_avg.loc['mean']
avg_temp
```

```python
#stworzenie nowej ramki danych z potrzebnymi danymi do daljszej części
avg_year_temp = pd.melt(my_data, id_vars=['Area', 'Months', 'Element', 'Unit'], var_name='Year', value_name='Temperature Change')
avg_year_temp
```

```python
# finalna ramka danych, która będzie wizualizowana na wykresach
avg_year_temp_melt = avg_year_temp.drop(['Months','Element','Unit'], axis=1)
avg_year_temp_melt = avg_year_temp_melt.rename(columns={'Area': 'Group'})
avg_year_temp_melt['Year'] = pd.to_numeric(avg_year_temp_melt['Year'])

final_data = avg_year_temp_melt
final_data
```

```python
# wyprowadzenie średniej zmiany temeratury dla wszystkich ID 

Temp_month = Temp_month.copy()
Temp_month.loc[:, "Suma"] = Temp_month.iloc[:, 5:63].sum(axis=1)
Temp_month["Suma"] = Temp_month["Suma"] / 59
```

```python
# wyświetlenie nowej ramki danych
top_5_avg = Temp_month
top_5_avg
```

```python
# ograniczenie ramkii danych do jedynie wymaganych kolumn
top_5_final = top_5_avg.loc[:, ["Area", "Suma"]]
top_5_final = top_5_final.rename(columns={'Suma': 'Temp_change'})
top_5_final
```

```python
# sortowanie malejąco
top_5_final_desc = top_5_final.sort_values("Temp_change", ascending=False)
top_5_final_desc
```

```python
top_5_final_desc = top_5_final_desc.iloc[:5, :]
plt.bar(top_5_final_desc["Area"], top_5_final_desc["Temp_change"], width=0.6)
plt.xlabel("Area")
plt.ylabel("Temperature change")
plt.title("Top 5 biggest temp change")
plt.show
```

```python
sns.barplot(x = 'Area', y = 'Temp_change', data = top_5_final_desc.iloc[:5, :], palette = 'flare', width=0.65)
plt.ylabel('Temperature change °C')
```
